## False Positives

The majority of false positives contain disaster-related vocabulary but do not describe real disaster events.

Examples include:

- metaphorical language
- jokes and sarcasm
- references to products, movies, or music
- historical discussions

The model tends to over-rely on specific keywords such as "earthquake", "explosion", or "mass murder".

## False Negatives

False negatives often describe disaster-related events indirectly.

Many tweets lack strong disaster keywords and instead rely on context.

Examples include:

- refugee crises
- hostage situations
- kidnappings
- humanitarian emergencies

These cases are difficult for bag-of-words models because the meaning is spread across the sentence rather than concentrated in individual words.

## Why Does TF-IDF Fail?

The TF-IDF representation ignores word order and context.

As a result:

1. Metaphorical uses of disaster-related words often trigger false alarms.
2. Real disasters described without explicit keywords are frequently missed.
3. Contextual relationships between words are not captured.

The model primarily relies on lexical signals rather than semantic understanding.


## Motivation for Transformer Models

The observed errors suggest that a context-aware language model may perform better.

Transformer architectures such as DistilBERT encode relationships between words and can distinguish between:

- literal vs figurative language
- event descriptions vs casual mentions
- semantic context surrounding disaster-related keywords
